```python
df_new = pd.DataFrame()

df_new['text'] = df['clean_text']
```

```python
df_new
```

```python
def truncate_text(text, max_tokens = 8000):
    return text[:max_tokens]
```

```python

def get_embeddings(text, model='text-embedding-3-small'):
    text = truncate_text(text)
    return client.embeddings.create(input = {text}, model= model).data[0].embedding
```

```python
question = 'what is a food allergy ?'
```

```python
question_embedding = get_embeddings(question)
```

```python
question, question_embedding
```

```python
import numpy as np
def fn( page_embedding):
    return np.dot(page_embedding, question_embedding)
```

```python
df_new['distance'] = df_new['embedding'].apply(fn)
```

```python
df_new.head()
```

```python
df_new.sort_values('distance', ascending=False, inplace=True)
```
